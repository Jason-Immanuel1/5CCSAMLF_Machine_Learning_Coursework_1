### Correlation Matrix

The only feature showing any correlation to 'outcome' is 'depth' suggesting a non-linear relationship between features.

```python
numeric_df = df[["outcome", "carat", "depth", "table", "price", "x", "y", "z"]]

sns.heatmap(data=numeric_df.corr(), annot=True, cmap="rainbow")
plt.title("Correlation Heatmap")
plt.savefig("Graphs/Correlation_Heatmap.png")
plt.show()
```

### Outcome Distribution
The 'outcome' variable shows a normal distribution with **mean:** -4.98 and **standard deviation:** 12.7.

```python
sns.displot(data=df["outcome"], kde=True, bins=20)
plt.title("Outcome Distribution")
plt.savefig("Graphs/Outcome_Distribution.png")
plt.show()
```

### Carat and Price Distribution
Both Carat and Price show log normal distributions.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data=df["carat"], kde=True, bins=20, ax=axes[0])
axes[0].set_title("Carat Distribution")

sns.histplot(data=df["price"], kde=True, bins=20, ax=axes[1])
axes[1].set_title("Price Distribution")

plt.savefig("Graphs/Carat_Price.png")
plt.show()
```

### Categorical Data Count

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 10))

sns.countplot(data=df["cut"], ax=axes[0])
axes[0].set_title("Cut Counts")

sns.countplot(data=df["color"], ax=axes[1])
axes[1].set_title("Color Distribution")

sns.countplot(data=df["clarity"], ax=axes[2])
axes[2].set_title("Clarity Distribution")

plt.savefig("Graphs/Cut_Color_Clarity_Distribution.png")
plt.show()
```

### Depth and Outcome Relation

```python
sns.scatterplot(x=df["depth"], y=df["outcome"], data=df)
plt.title("Depth-Outcome Relationship")
plt.savefig("Graphs/Depth_Outcome_Relationship.png")
plt.show()
```

## Conclusion
***

After analyzing the synthetic Diamonds dataset, it is evident that there is little to no linear correlation between the features and the target variable (other than 'depth'). Therefore, a tree based model such as a Gradient Boosting Machine (GBM) or Random Forest is most suitable for predicting the ‘outcome’. Linear models will struggle to capture the complex, non-linear interactions present in the data. In contrast, GBMs and other ensemble tree methods will automatically identify and model these relationships, resulting in superior predictive performance on non-linear tabular datasets.
